## Menghilangkan atau menyingkirkan Missing Value

### Hilangkan kolom yang mempunyai missing value tidak dapat diisi dengan value lain secara logis (seperti nama, host_id, dll)

```python
airbnb =  data.loc[:, ['id', 'host_id', 'neighbourhood_group',
       'neighbourhood', 'latitude', 'longitude', 'room_type', 'price',
       'minimum_nights', 'number_of_reviews', 'last_review',
       'reviews_per_month', 'calculated_host_listings_count',
       'availability_365']]
airbnb
```

```python
data.iloc[:, 4:16]
```

### Isi missing value pada kolom "last_review" dengan tanggal paling sering muncul"

Ubah tipe data pada kolom "last_review" menjadi `date`

```python
data
```

```python
data.dtypes
```

```python
# Konversi data ke format penanggalan
airbnb["last_review"] = pd.to_datetime(airbnb["last_review"], format='%Y-%m-%d', errors = 'coerce')
airbnb["last_review"]
```

```python
airbnb.dtypes
```

```python
airbnb['last_review'] = airbnb['last_review'].fillna(airbnb['last_review'].mode()[0])
airbnb
```

```python
airbnb.info()
```

### Isi missing value pada kolom reviews_per_month dengan `Mean`

```python
airbnb['reviews_per_month'].fillna(airbnb['reviews_per_month'].mean())
airbnb
```

```python
airbnb.info()
```

## Buat kategori harga berdasarkan kolom "price"

```python
airbnb["price"].value_counts()
```

```python
rules = [(airbnb["price"] <= 1000), ((data["price"] > 1000)&(data["price"] <= 6000 )), (data["price"] <= 10000)] 
info = ["Cheap", "Medium", "Expensive"]
airbnb["Cheap", "Medium", "Expensive"] = np.select(rules, info)
airbnb.head(20)
```

```python
airbnb.columns = ['id', 'host_id', 'neighbourhood_group', 'neighbourhood', 'latitude', 'longitude', 'room_type', 'price',
                  'minimum_nights','number_of_reviews', 'last_review', 'reviews_per_month', 'calculated_host_listings_count', 'availability_365',
                  'range_price']
airbnb
```

## Visualisasikan kolom "price" untuk membandingkan harga kamar yang tersedia

```python
airbnb["room_type"].value_counts()
```

```python
airbnb["range_price"].value_counts()
```

```python
airbnb["price"].value_counts()
```

```python
d = airbnb.groupby('room_type')['price'].mean().sort_values(ascending=False)
fig, ax= plt.subplots(figsize=(15,5))
sns.barplot(d.index.tolist(), d.values, ax=ax)
ax.set_title('Average price of room types')
fig.show()
```

```python
d = airbnb.groupby('room_type')['price'].mean().sort_values(ascending=False)
fig, ax= plt.subplots(figsize=(15,7))
sns.barplot(d.index.tolist(), d.values, ax=ax)
ax.set_title('Average price of room types')
fig.show()
```

```python
fig, ax= plt.subplots(2,1, figsize=(20,10))

ax[0].set_title("Histogram of prices (total)")
sns.distplot(data['price'], kde=True, ax=ax[0])
ax[1].set_title("Histogram of prices (< $800)")
sns.distplot(data[data['price'] < 800]['price'], kde=True, ax=ax[1])



fig.suptitle("Distribution of prices")
fig.show()
```
