## 4. Conditionally Adding Column

```python
import numpy as np
```

```python
data["DISTANCE"] = np.where(data["MILES*"] > 5, "Long Trip", "Short Trip")
```

```python
data.head()
```

```python
data["DISTANCE"].value_counts()
```

```python
data["YEAR"] = np.array("2020")
data
```

```python
data["TIME_CAT"] = np.where(data["START_DATE*"] > "1/5/2016", "New Trip", "Old Trip")
data
```

## Latihan 5

### Soal 1 Buat kolom baru dengan kondisi berikut: 
    a) >10 : Long Trip
    b) 5-10 : Medium trip
    c) <5 : Short Trip

```python
rules = [(data["MILES*"] > 10), ((data["MILES*"] < 10)&(data["MILES*"] >= 5)), (data["MILES*"] < 5)] 
info = ["Long Trip", "Medium Trip", "Short Trip"]
data["Long Trip", "Medium Trip", "Short Trip"] = np.select(rules, info)
data.head(10)
```

# 5. Groupby Summarize

```python
data.groupby("START*")["MILES*"].agg(["mean", "sum"]).head()
```
